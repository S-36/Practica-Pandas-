## Filtrar y Segmentar datos 

Se establece un filtro para mostrar resultados de una columna especifica junto con los resultados esperados del filtro 
En este codigo se busca cuales son los datos de la columna "Estado" que tienen el valor de "Entregado"
Ademas de que se puede unir con el ".describe()" para ver una version mas resumida e incluso con mas contexto de todo lo que 
incluyen estos **Estados con resultados de Entregado**

```python
# Solo envíos entregados
entregados = df[df["Estado"] == "Entregado"]
print(entregados.describe())
```

Tambien con Tablas que contengan datos **INT** se pueden filtrar con condicionales, ademas de mostrar que se pueden combinar varias de estas funciones por ejemplo, para solo conseguir el numero de tarifas que son mayor a 500

```python
# Envíos con tarifa mayor a $500
caros = df[df["Tarifa_USD"] > 500].value_counts().count()
print(caros)
```

Lo mismo para Tablas con que tengan valores de **str** que de igual manera, recordando que el **.describe()** ignora las columnas de su resultado con texto, solo mantiene las de numero por lo que se tendran que manejar aparte

```python
# Envíos de un cliente específico
logicargo = df[df["Cliente"] == "LogiCargo SA"].describe()
print(logicargo)
```

La combinacion de condicionales tambien es necesaria para hacer un mayor filtro de los datos que estamos buscando 

```python
# Combinar condiciones
retrasados_pesados = df[(df["Estado"] == "Retrasado") & (df["Peso_kg"] > 10000)]
print(f'La cantidad de camiones retrasados de una tonelada son: {retrasados_pesados.value_counts().count()}')
print(retrasados_pesados.describe())
```

## Agrupar y calcular metricas 

Para estas metricas es necesario tener un buen entendimiento de los datos que se van a juntar 

```python
# Ingresos totales por cliente
ingresos_por_cliente = df.groupby("Cliente")["Tarifa_USD"].sum().sort_values(ascending=False)
print(ingresos_por_cliente)
```

```python

# Envíos y tarifa promedio por estado
resumen_estado = df.groupby("Estado").agg(
    total_envios=("ID_Envio", "count"),
    tarifa_promedio=("Tarifa_USD", "mean"),
    ingreso_total=("Tarifa_USD", "sum")
)
print(resumen_estado)
```

```python
# Distancia promedio por ruta
rutas = df.groupby(["Origen", "Destino"])["Distancia_km"].mean()
print(rutas)
```
